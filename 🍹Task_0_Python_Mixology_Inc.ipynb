### üç∏ Task 0.1 - Weather-Based Drink Recommender
**Concepts: if-elif-else**

Complete the function that recommends a cocktail based on the weather.

Example input and output:

suggest_drink("hot") ‚Üí "Try a Mojito!"

suggest_drink("cold") ‚Üí "Go for a Hot Toddy!"

suggest_drink("rainy") ‚Üí "How about a Dark 'n' Stormy?"

Your code below:

```python
def suggest_drink(weather):
  #complete the function
  if(weather=="hot"):
     print("Try a Mojito!")
  elif(weather=="cold"):
     print("Go for a Hot Toddy!")
  elif(weather=="rainy"):
     print("How about a Dark 'n' Stormy?")
  else:
     print("Something special on your mind?")
  pass

suggest_drink("rainy")
```

###ü•Ç Task 0.3 - Cocktail Menu Operations
**Concepts: dictionaries, .get(), .items(), sorting with lambda**

You are given a cocktail menu as a dictionary with cocktail names as keys and their ratings (out of 5) as values.

Follow the instructions and write relevant code.

```python
menu = {
    "Mojito": 4.5,
    "Martini": 4.8,
    "Negroni": 4.2
}

# Print all cocktails with their ratings.
for cocktail, rating in menu.items():
  print(f"Cocktail: {cocktail} - Rating: {rating} ‚≠ê")

# Check if Negroni is on the menu. If yes, return rating, if not, return a default message "Sorry! We don't serve this.".
if "Negroni" in menu:
  print(f"Rating: ",menu["Negroni"])
else:
  print("Sorry! We don't serve this.")

# Check if LIIT is on the menu. If yes, return rating, if not, return a default message "Sorry! We don't serve this.".
if "LIIT" in menu:
  print(f"Rating: ",menu["LIIT"])
else:
  print("Sorry! We don't serve this.")

# Add a new drink: "Pina Colada" with a rating of 4.6.
menu["Pina Colada"] = 4.6
print("New drink added!")

# Sort the cocktails from highest to lowest rated and print them in order.
sort_menu = sorted(menu.items(), key=lambda x:x[1], reverse=True)
print("Cocktails based on ratings(highest to lowest):")
for cockail,rating in sort_menu:
  print(f"{cocktail}: {rating}")
```

###üç∑ Task 0.4 - Bartender Profile Generator
**Concepts: class, init, method, datetime, string formatting**

Complete a class Bartender with the following:

**Attributes:**

name (string)

joining_month (string, e.g. "February")

favorite_cocktail (string)

**Methods:**

work_duration(): calculates how many *full* months the bartender has worked at Mixology Inc. till present.

introduce(): returns a string like:
‚ÄúHi, I‚Äôm Aarushi. I‚Äôve been working here for 5 months and my signature drink is the Cosmopolitan.‚Äù

```python
from posixpath import join
import datetime

class Bartender:
  def __init__(self, name, joining_month, favorite_cocktail):
    # complete the constructor
    self.name = name
    self.joining_month = joining_month
    self.favorite_cocktail = favorite_cocktail
    pass

  def work_duration(self):
    # calculate how many full months since joining
    current_month = datetime.datetime.now().month
    joining_month = datetime.datetime.strptime(self.joining_month,"%B").month
    duration = current_month - joining_month
    if duration<0:
      duration+=12
    return duration

  def introduce(self):
    # return the formatted string
    worked_for_months = self.work_duration()
    print(f"Hi, I'm {self.name}. I‚Äôve been working here for {worked_for_months} months and my signature drink is the {self.favorite_cocktail}.")
    pass


b = Bartender("Aarushi", "February", "Cosmopolitan")
b.introduce()
```
